```python
import os
import shutil
import ast
import itertools
from collections import Counter

import gym
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.metrics import  ConfusionMatrixDisplay
from sklearn.metrics import f1_score

from stable_baselines3 import PPO, A2C
from stable_baselines3.common.monitor import Monitor
from stable_baselines3.common.results_plotter import load_results, ts2xy
from stable_baselines3.common.callbacks import BaseCallback, CheckpointCallback
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.env_util import make_vec_env

from pcwcf.common.policies import ConstrainedActorCriticPolicy
from gym_pcwcf.config.config_cicids2017_webattacks import FEATURES
```

```python
eval_log_dir = "/tmp/gym/val/"
```

```python
def create_dir(dirname):
    # Create log dir     
    try:
        shutil.rmtree(eval_log_dir)
    except:
        pass
    finally:
        os.makedirs(eval_log_dir, exist_ok=True)
```

```python
def evaluate(lambda_coefficient = 0.01,
            number = '1',
            results=[],
            mode='TEST'):
    create_dir(eval_log_dir + number +"/")
    env_id = 'gym_pcwcf:cwcf-v0'
    env_kwargs = dict(mode=mode,
                 is_binary_classification=True,
                 random_mode=False,
                 lambda_coefficient=lambda_coefficient,
                 terminal_reward=[[0, -0.3], 
                                  [-0.7, 0]])
    monitor_kwargs = dict(info_keywords=('y_true', 'y_predict', 'features'))
    eval_env = gym.make(env_id, **env_kwargs)
    eval_env = Monitor(eval_env, filename=eval_log_dir, **monitor_kwargs)
    model_dir = '/home/student/projects/site-packages/rl-baselines3-zoo/logs/11_01_2022/ppo/cwcf-v0_' +  number +'/best_model.zip'
    model = PPO.load(model_dir)
    mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=len(eval_env.data_y))
    df = load_results(eval_log_dir)
    accuracy = np.mean(df.y_true.values==df.y_predict.values)
    mean_reward1 = np.mean(df.r)
    used_features = np.mean(df.l)-1
    f1score = f1_score(df.y_true, df.y_predict)
    results.append([mean_reward, accuracy,f1score, used_features])
```

```python
lambdas = np.concatenate([np.repeat(lmbd, 5) for lmbd in [0.1, 0.05, 0.01, 0.005, 0.001, 0.0005, 0.0001] ])
```

```python
results_val = []
for i in np.arange(1,len(lambdas)+1):
    evaluate(lambda_coefficient = lambdas[i-1], number = str(i) , results=results_val, mode='VAL')
arr_val = np.array(results_val)
plt.scatter(arr_val[:,3], arr_val[:,2])
plt.xlabel('Среднее количество используемых признаков')
plt.ylabel('F мера VALIDATION')
plt.rcParams.update({'font.size': 18})
plt.show()
```

```python
results_test = []
for i in np.arange(1,len(lambdas)+1):
    evaluate(lambda_coefficient = lambdas[i-1], number = str(i) , results=results_test, mode='TEST')
arr_test = np.array(results_test)
plt.scatter(arr_test[:,3], arr_test[:,2])
plt.xlabel('Среднее количество используемых признаков')
plt.ylabel('F мера')
plt.rcParams.update({'font.size': 18})
plt.show()
```

```python
results_real = []
for i in np.arange(1,len(lambdas)+1):
    evaluate(lambda_coefficient = lambdas[i-1], number = str(i) , results = results_real, mode='REAL')
arr = np.array(results_real)
plt.scatter(arr[:,3], arr[:,2])
plt.xlabel('Среднее количество используемых признаков')
plt.ylabel('F мера')
plt.rcParams.update({'font.size': 18})
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score
def plot_results(index):
    lambda_coefficient = lambdas[index]
    number = str(index + 1)
    create_dir(eval_log_dir + number +"/")
    env_id = 'gym_pcwcf:cwcf-v0'
    env_kwargs = dict(mode='TEST',
                 is_binary_classification=True,
                 random_mode=False,
                 lambda_coefficient=lambda_coefficient,
                 terminal_reward=[[0, -0.3], 
                                  [-0.7, 0]])
    monitor_kwargs = dict(info_keywords=('y_true', 'y_predict', 'features'))
    eval_env = gym.make(env_id, **env_kwargs)
    eval_env = Monitor(eval_env, filename=eval_log_dir, **monitor_kwargs)
    model_dir = '/home/student/projects/site-packages/rl-baselines3-zoo/logs/11_01_2022/ppo/cwcf-v0_' +  number +'/best_model.zip'
    model = PPO.load(model_dir)
    mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=len(eval_env.data_y))
    df = load_results(eval_log_dir)
    accuracy = np.mean(df.y_true.values==df.y_predict.values)
    mean_reward1 = np.mean(df.r)
    used_features = np.mean(df.l)-1
    f1score = f1_score(df.y_true, df.y_predict)
    disp = ConfusionMatrixDisplay.from_predictions(df.y_true, df.y_predict)
    disp.figure_.suptitle("Матрица неточностей")
    print(classification_report(df.y_true, df.y_predict))
    plt.rcParams.update({'font.size': 12})
    plt.show()
    plt.hist(df.l.map(lambda x: x-1), bins=10, align='left')
    plt.title('Гистограмма количества использованных признаков\nlambda=' + str(lambda_coefficient))
    plt.rcParams.update({'font.size': 12})
    plt.show()
    
    features_union = list(itertools.chain.from_iterable(df.features.map(ast.literal_eval)))
    features_counts = Counter(features_union)
    n = 20
    y = [c for f, c in features_counts.most_common(n)]
    x = [FEATURES[f-2] for f, c in features_counts.most_common(n)]
    plt.bar(x,y)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.title('Частота использования признаков признаков\nlambda=' + str(lambda_coefficient))
    plt.rcParams.update({'font.size': 12})
    plt.show()
    detailed_results.append([precision_score(df.y_true, df.y_predict), recall_score(df.y_true, df.y_predict), f1_score(df.y_true, df.y_predict), used_features])
```
